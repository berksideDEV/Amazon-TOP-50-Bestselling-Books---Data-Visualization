```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
amazon = pd.read_csv("/content/bestsellers with categories.csv")
amazon.head(10)
```

```python

amazon['Name'] = amazon['Name'].astype("category")
amazon['Author'] = amazon['Author'].astype("category")
amazon['Genre'] = amazon['Genre'].astype("category")
```

```python
# Kullanıcı puanlaması bazlı, kitap türlerinin sayısını inceleyiniz.

#Cevap
#4.0 - 4.9 değerlendirme puanı aralığında türü kurgu olmayan kitapların daha fazla olduğunu söyleyebiliriz.
sns.histplot(x="User Rating",hue="Genre",multiple="stack", data=amazon)
sns.set(rc = {'figure.figsize':(12,8)})
```

```python
#2.Soru
#Yıla göre türü kurgu ve kurgu olmayan kitapların fiyatlarını inceleyin.

#Cevap
#Bu grafikte ise tarihe göre kurgu ve kurgu olmayan kitapların fiyat karşılaştırmalarını yapabiliriz.
#Örneğin, 2010 yılında kurgu olmayan kitapların, kurgu olanlara göre daha pahalı olduğunu söyleyebiliriz.

sns.catplot(x = "Genre", y= "Price",hue="Year", kind="bar", data=amazon)
```

```python
#3. Soru
#2009 ile 2019 yılları arasında hangi kitap türü daha fazla piyasaya çıkarılmıştır ?

#Cevap
#2009 ile 2019 yılları aralığında türü kurgu olmayan kitapların, türü kurgu olanlara göre piyasaya daha çok çıkarıldığını söyleyebiliriz.

sns.countplot(x="Year",hue="Genre",linewidth=2,data=amazon)
```

```python
#4.SORU
#Yıl bazında Hangi kitap türü daha çok değerlendirilmiştir ?

#Cevap : Grafiği incelediğimiz zaman 2010-2016 yılları arasında kurgu türünde incelenme sayısının kurgu olmayan kitap türüne göre daha fazla incelendiğini görüyoruz.
a = sns.lineplot(y="Reviews", x="Year", hue="Genre" ,data=amazon,linewidth=3);
a.set(title = "Yıl bazında incelenen kitap türleri")
```

```python
#5. Soru

#Kitap fiyatı bazında kitap türlerinin değerlendirme sayısını yorumlayınız.

#Cevap
#Çoğunlukla 0-20 fiyat aralığında olan kitaplar, 0-20.0000 arasında incelenmiştir.
#0-20 fiyat aralığında en fazla incelenme sayısına sahip kitap türü ise kurgu türüdür.

sns.scatterplot(x='Price',y='Reviews',hue='Genre',style='Genre',data=amazon)
```

```python
#6. SORU
#Kullanıcı puanlamasının değerlendirme sayısı ile ilişkisini inceleyerek hangi kitap türlerinin aykırı değerlere sahip olduğunu gösteriniz.

# Kullanıcı puanlaması 4.8 olan kısımda 80.000 den fazla inceleme sayısına sahip, kurgu türünde yer alan aykırı bir değer bulunmaktadır.
# Bir diğer aykırı değer ise, Kullanıcı puanlaması 4.1 olan kısımda ortalama 80.000 inceleme sayısına sahip, kurgu türünde yer alan değerdir.
# En çok puan alan kısım yani kullanıcı puanlaması 4.9 olan bölümde, kurgu türünde yer alan kitaplar kurgu olmayanlara göre inceleme sayısı bakımından daha fazla dağılım göstermişlerdir.
plt.figure(figsize=(14,8))
sns.scatterplot(x="User Rating",y="Reviews",data=amazon,hue="Genre")
```
